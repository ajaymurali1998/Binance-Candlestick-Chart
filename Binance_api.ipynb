# Plotly Graph

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
df=pd.read_csv('eth_candle.csv')
df.head()
```

## CandleStick Chart using Plotly

The candlestick chart is a style of financial chart describing open, high, low and close for a given x coordinate (most likely time). 

The boxes represent the spread between the open and close values and the lines represent the spread between the low and high values. 

Sample points where the close value is higher(lower) then the open value are called increasing (decreasing).

By default, increasing candles are drawn in green whereas decreasing are drawn in red.

We are also calculating moving average metrics to better understand the price.

```python
## At first we want to create a column for calculating 5 day moving average
df['5dma']=df['close'].rolling(window=5).mean()

## creating candlestick chart  
fig=go.Figure(data=[go.Candlestick(x=df['opentime'],
                                   open=df['open'],
                                   high=df['high'],
                                   low=df['low'],
                                   close=df['close'])])

## adding the 5 day MA scatter plot in figure
fig.add_trace(
    go.Scatter(
        x=df['opentime'],
        y=df['5dma'],
        line=dict(color='#e0e0e0'),
        name="5 day MA"))

# removing rangeslider
fig.update_layout(xaxis_rangeslider_visible=False,template='plotly_dark')



fig.update_yaxes(type='log')
# adding titles
fig.update_layout(title='Binance ETH/BTC stock price analysis',
                  yaxis_title='Ethereum price in Bitcoin(BTC)',
                   xaxis_title='Date')

fig.show()
```
